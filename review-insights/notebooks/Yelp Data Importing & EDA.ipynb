```python
# IMPORT NECESSARY LIBRARIES AND DATA
import pandas as pd
import numpy as np
import os

# Import the entire review JSON file
review_file_path = 'yelp_academic_dataset_review.json'
review_df = pd.read_json(review_file_path, lines=True)
print(f"Successfully imported the entire review file with shape: {review_df.shape}")

# Print the size of the review file
review_file_size = os.path.getsize(review_file_path)
print(f"Size of the review file: {review_file_size} bytes")

# Import the business JSON file
business_file_path = 'yelp_academic_dataset_business.json'
business_df = pd.read_json(business_file_path, lines=True)
print(f"Successfully imported the entire business file with shape: {business_df.shape}")

# Print the size of the business file
business_file_size = os.path.getsize(business_file_path)
print(f"Size of the business file: {business_file_size} bytes")

# Import the tips JSON file
tip_file_path = 'yelp_academic_dataset_tip.json'
tip_df = pd.read_json(tip_file_path, lines=True)
print(f"Successfully imported the entire tip file with shape: {tip_df.shape}")

# Print the size of the tips file
tip_file_size = os.path.getsize(tip_file_path)
print(f"Size of the tip file: {tip_file_size} bytes")

# Import the user JSON file
user_file_path = 'yelp_academic_dataset_user.json'
user_df = pd.read_json(user_file_path, lines=True)
print(f"Successfully imported the entire user file with shape: {user_df.shape}")

# Print the size of the user file
user_file_size = os.path.getsize(user_file_path)
print(f"Size of the user file: {user_file_size} bytes")

# Import the checkin JSON file
checkin_file_path = 'yelp_academic_dataset_checkin.json'
checkin_df = pd.read_json(checkin_file_path, lines=True)
print(f"Successfully imported the entire checkin file with shape: {checkin_df.shape}")

# Print the size of the checkin file
checkin_file_size = os.path.getsize(checkin_file_path)
print(f"Size of the checkin file: {checkin_file_size} bytes")
```

```python
# Total number of businesses in Philadelphia

philly_busi = business_df[business_df['city']=='Philadelphia']
print(f"Number of businesses in Philadelphia: {philly_busi.shape[0]}")

# Categories of businesses in Philadelphia
philly_busi_cat = philly_busi['categories'].explode().value_counts()
print("Top 30 business categories in Philadelphia:")
print(philly_busi_cat.head(30))
```

```python
# Create a subset of business categories related to food

# Food related categories
food_cats = ['Restaurants', 'Food', 'Pizza', 'Burgers', 'Sandwiches', 'Coffee', 'Tea', 
    'Bakeries', 'Desserts', 'Ice Cream', 'Ice Cream & Frozen Yogurt', 'Cafes', 'Fast Food',
    'Pubs', 'Breweries', 'Wineries', 'Caterers', 'Beer', 'Wine & Spirits', 'Food Trucks', 
    'Coffee & Tea', 'Mexican', 'Bistros', 'Buffets', 'Cajun/Creole', 'Caribbean', 'French',
    'German', 'Irish', 'Mediterranean', 'Middle Eastern', 'Salad', 'Seafood', 'Steakhouses', 
    'Sushi Bars', 'Tapas Bars', 'Vegetarian', 'Vegan', 'Breakfast & Brunch', 'Gluten-Free', 
    'Halal', 'Kosher', 'Korean', 'Japanese', 'Chinese', 'Thai', 'Indian', 'Vietnamese',
    'African', 'Turkish', 'Russian', 'Spanish', 'Portuguese', 'British', 'Eastern European', 
    'Latin American', 'Argentine', 'Brazilian', 'Peruvian', 'Fried Chicken', 'Hot Dogs', 
    'Juice Bars & Smoothies', 'Ramen', 'Noodles', 'Dim Sum', 'Belgian', 'Cuban', 'Ethiopian', 
    'Lebanese', 'Moroccan', 'Pakistani', 'Sri Lankan', 'Afghan', 'Cambodian', 'Indonesian', 
    'Malaysian', 'Mongolian', 'Colombian', 'Nepalese', 'Singaporean', 'Somali', 'Taiwanese', 
    'Tibetan', 'Uzbek', 'Yemeni', 'American (Traditional)', 'American (New)', 'Barbecue', 
    'Delis', 'Diners', 'Gastropubs', 'Italian', 'Greek', 'American', 'Bars', 'Bagels', 
    'Soup', 'Juice Bars', 'Smoothies', 'Donuts', 'Chicken Wings', 'BBQ', 'Barbeque', 'Sushi', 
    'Wine Bars', 'Cocktail Bars', 'Specialty Food', 'Ethnic Food', 'Asian Fusion', 'Filipino']

# Function to check if business has food-related categories
def is_food_related(categories):
    if pd.isna(categories):  # Handle missing values
        return False
    
    # Split categories into individual items and check each one
    cat_list = [cat.strip() for cat in str(categories).split(',')]
    return any(keyword.lower() in cat.lower() for cat in cat_list for keyword in food_cats)

# Filter Philadelphia businesses to food-related only
philly_food = philly_busi[philly_busi['categories'].apply(is_food_related)]

print(f"Total Philadelphia businesses: {philly_busi.shape[0]}")
print(f"Food-related businesses: {philly_food.shape[0]}")
print(f"Percentage: {(philly_food.shape[0] / philly_busi.shape[0]) * 100:.1f}%")

# Total number of businesses that are food-related
print(f"Number of food-related businesses in Philadelphia: {philly_food.shape[0]}")

# Split categories into individual items for counting
philly_food_cats = philly_food['categories'].str.split(', ').explode().str.strip()

# Check the top categories in filtered dataset
print("\nTop 30 food related categories in Philadelphia:")
print(philly_food_cats.value_counts().head(30))
```

```python
# Philadelphia food businesses
print(f"Philadelphia food businesses: {philly_food.shape[0]:,}")

# Filter reviews to match Philadelphia food business IDs
philly_food_ids = philly_food['business_id'].unique()
philly_food_reviews = review_df[review_df['business_id'].isin(philly_food_ids)]
print(f"Matching reviews: {philly_food_reviews.shape[0]:,}")

# Filter tips to match Philadelphia food business IDs (only keeping business_id and text)
philly_food_tips = tip_df[tip_df['business_id'].isin(philly_food_ids)][['business_id', 'text']]
print(f"Matching tips: {philly_food_tips.shape[0]:,}")

# Filter checkins to match Philadelphia food business IDs (only keeping business_id and date)
philly_food_checkins = checkin_df[checkin_df['business_id'].isin(philly_food_ids)][['business_id', 'date']]
print(f"Matching checkins: {philly_food_checkins.shape[0]:,}")

# Filter users to match those who wrote reviews for Philadelphia food businesses
philly_user_ids = philly_food_reviews['user_id'].unique()
philly_food_users = user_df[user_df['user_id'].isin(philly_user_ids)]
print(f"Matching users: {philly_food_users.shape[0]:,}")

# Merge reviews with businesses
philly_combined = philly_food_reviews.merge(
    philly_food,
    on='business_id',
    how='left',
    suffixes=('_review', '_business')
)
print(f"\nAfter merging reviews + businesses: {philly_combined.shape}")

# Aggregate tips per business to avoid cartesian product
philly_food_tips = philly_food_tips.rename(columns={'text': 'tip_text'})
tips_aggregated = philly_food_tips.groupby('business_id').agg({
    'tip_text': lambda x: ' | '.join(x.dropna())
}).reset_index()
tips_aggregated['tip_count'] = philly_food_tips.groupby('business_id').size().values

# Merge tips
philly_combined = philly_combined.merge(
    tips_aggregated,
    on='business_id',
    how='left'
)
print(f"After merging tips: {philly_combined.shape}")

# Aggregate checkins per business
philly_food_checkins = philly_food_checkins.rename(columns={'date': 'checkin_dates'})
checkins_aggregated = philly_food_checkins.groupby('business_id').agg({
    'checkin_dates': lambda x: ' | '.join(x.dropna())
}).reset_index()
checkins_aggregated['checkin_count'] = philly_food_checkins.groupby('business_id').size().values

# Merge checkins
philly_combined = philly_combined.merge(
    checkins_aggregated,
    on='business_id',
    how='left'
)
print(f"After merging checkins: {philly_combined.shape}")

# Merge users with reviews
philly_final = philly_combined.merge(
    philly_food_users,
    on='user_id',
    how='left',
    suffixes=('', '_user')
)
print(f"\nFinal dataset: {philly_final.shape}")
print(f"\nFinal dataset columns: {philly_final.columns.tolist()}")
print(f"\nSample of data:")
print(philly_final[['business_id', 'name', 'user_id', 'tip_text', 'tip_count', 'checkin_count']].head())
```
